DFT Plot of dB Spectrum for hamming window of lenth t = 5ms,10ms,20ms,40ms for F0 120 Hz

```python
import numpy as np
from numpy import pi, cos, log10
from scipy import signal
from matplotlib import pyplot as plt
from scipy.fftpack import fft, fftshift
from scipy.io import wavfile
from scipy.fft import fftfreq
fs, data1 = wavfile.read('/content/u_F0-120_F1-300F2-870F3-2240_B1-50B2B3-100_Fs16k.wav')

x1 = data1[int(0.20*fs):int((0.20+0.005)*fs)] #generating frame signal
timeaxis = np.linspace(0,0.005,len(x1))
t1 =0.005
L1 = int(t1*fs) #l = duration of window*fs
w1 = x1*signal.hamming(L1)
#zero padding
N=1024
T=1/16000
y1 = np.concatenate([w1, np.zeros(N-L1)])
z1 = fft(y1,1024)
magX1 = abs(z1[0:N//2])
spec1 = 20*log10(magX1)
plt.figure(1)
plt.plot(np.arange(0,(fs/2),fs/float(N)), spec1)
plt.title('Hamming Window Size 5ms for F0 = 120 Hz')
plt.xlabel('freq in Hz')
plt.ylabel('Magnitude in dB')
print( 'measured formant frequency' + str(np.argmax(z1))+ 'Hz' )

#window length t = 10ms
x2 = data1[int(0.20*fs):int((0.20+0.01)*fs)] #generating frame signal
timeaxis = np.linspace(0,0.01,len(x2))
t2 =0.01
L2 = int(t2*fs) #l = duration of window*fs
w2 = x2*signal.hamming(L2)
#zero padding
N=1024
T=1/16000
y2 = np.concatenate([w2, np.zeros(N-L2)])
z2 = fft(y2,1024)
magX2 = abs(z2[0:N//2])
spec2 = 20*log10(magX2)
plt.figure(2)
plt.plot(np.arange(0,(fs/2),fs/float(N)), spec2)
plt.title('Hamming Window Size 10ms for F0 = 120 Hz')
plt.xlabel('freq in Hz')
plt.ylabel('Magnitude in dB')

#window length t = 20ms
x3 = data1[int(0.20*fs):int((0.20+0.02)*fs)] #generating frame signal
timeaxis = np.linspace(0,0.02,len(x3))
t3 =0.02
L3 = int(t3*fs) #l = duration of window*fs
w3 = x3*signal.hamming(L3)
#zero padding
N=1024
T=1/16000
y3 = np.concatenate([w3, np.zeros(N-L3)])
z3 = fft(y3,1024)
magX3 = abs(z3[0:N//2])
spec3 = 20*log10(magX3)
plt.figure(3)
plt.plot(np.arange(0,(fs/2),fs/float(N)), spec3)
plt.title('Hamming Window Size 20ms for F0 = 120 Hz')
plt.xlabel('freq in Hz')
plt.ylabel('Magnitude in dB')

#window length t = 40ms
x4 = data1[int(0.20*fs):int((0.20+0.04)*fs)] #generating frame signal
timeaxis = np.linspace(0,0.04,len(x4))
t4 =0.04
L4 = int(t4*fs) #l = duration of window*fs
w4 = x4*signal.hamming(L4)
#zero padding
N=1024
T=1/16000
y4 = np.concatenate([w4, np.zeros(N-L4)])
z4 = fft(y4,1024)
magX4 = abs(z4[0:N//2])
spec4 = 20*log10(magX4)
plt.figure(4)
plt.plot(np.arange(0,(fs/2),fs/float(N)), spec4)
plt.title('Hamming Window Size 40ms for F0 = 120 Hz')
plt.xlabel('freq in Hz')
plt.ylabel('Magnitude in dB')
```

From the above Plots for frequency f0 =120Hz and hamming window we can see as the window size increases Magnitude spectrum becomes more dense i.e no. of peaks increases.

Pitch period for window length 5ms = 2/1000 =0.002

Pitch period for window length 10ms = 3/1000 =0.003

Pitch period for window length 5ms = 8/1000 =0.008

Pitch period for window length 5ms = 10/1000 =0.01

with increase in window length pitch period increases
